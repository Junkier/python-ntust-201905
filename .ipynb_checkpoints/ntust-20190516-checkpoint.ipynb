```python
### Raw

mathStudScores = [
    {"name":"Jeff", "score":95},    
    {"name":"Leo" , "score":63},
    {"name":"Keven", "score":100},
    {"name":"Apple", "score":77},
    {"name":"Elle", "score":84},
    {"name":"Emma", "score":60},
]

engStudScores = [
    {"name":"Jeff","score":58},    
    {"name":"Leo", "score":100},
    {"name":"Keven","score":97},
    {"name":"Apple","score":85},
    {"name":"Elle","score":92},
    {"name":"Emma","score":41},
]

# 數學 : 總合
# 英文 : 平均 & 不及格名單

###
# Definition
### Math
def getScoresSum(data):
    # result : 總合
    result = 0
    
    for ele in data:
        result += ele["score"]
        
    return result

### Eng
def getScoreAvg(data):
    result = 0
    for ele in data:
        result += ele["score"]
    return result / len(data)


def getFailedList(data):
    result = []
    for ele in data:
        if ele["score"] < 60 :
            result.append(ele["name"])
    return result
    
###
# Main
    
mathSum = getScoresSum(mathStudScores)    
print("Math Sum :", mathSum)

mathAvg = getScoreAvg(mathStudScores)    
print("Math Avg :", mathAvg)

engAvg  = getScoreAvg(engStudScores)
print("Eng Avg :",engAvg)

failedList = getFailedList(engStudScores)
print("Eng failed list : ",failedList)

```

```python
class Person:
    city = "Taipei" 
    
    def __init__(self,name,age,location = Person.city):
        self.name = name
        self.age  = age
        self.location = location
        
    def sayMyName(self):
        print( "HiHi , I'm {}".format(self.name) )
        
    def showMyAge(self):
        print("I am {} years old.".format(self.age) )
        

## 物件 = 類別 產生()
person1 = Person("Jeff",18)
person2 = Person("Leo",22,"New Taipei City")

print(person1.name)
print(person2.age)

print("-"*80)

person1.sayMyName()
person2.sayMyName()

print("-"*80)

person1.showMyAge()
person2.showMyAge()

print("-"*80)

print(person1.location)
print(person2.location)


```

```python
class MathClass:
    def __init__(self,data):
        self.data = data
    
    def getScoresAvg(self):
        result = 0
        for ele in self.data:
            result += ele["score"]
        return result / len(self.data)
        
class EngClass:
    def __init__(self,data):
        self.data = data
    
    def getScoresAvg(self):
        result = 0
        for ele in self.data:
            result += ele["score"]
        return result / len(self.data) 
    
    def getFailedList(self):
        result = []
        for ele in self.data:
            if ele["score"]<60:
                result.append(ele["name"])
        return result
    
mathStudScores = [
    {"name":"Jeff", "score":95},    
    {"name":"Leo" , "score":63},
    {"name":"Keven", "score":100},
    {"name":"Apple", "score":77},
    {"name":"Elle", "score":84},
    {"name":"Emma", "score":60},
]

engStudScores = [
    {"name":"Jeff","score":58},    
    {"name":"Leo", "score":100},
    {"name":"Keven","score":97},
    {"name":"Apple","score":85},
    {"name":"Elle","score":92},
    {"name":"Emma","score":41},
]      

mathClass = MathClass(mathStudScores)
print("Math Avg : ", mathClass.getScoresAvg())


engClass  = EngClass(engStudScores)
print("Eng Avg : ",  engClass.getScoresAvg())
print("Failed list :" , engClass.getFailedList())
```
