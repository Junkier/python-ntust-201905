```python
# trTags = soup.select("table tr")
trTags = soup.select("table tr.trip-column")

dataList = []

for tr in trTags:
#     print(tr)
#     print(tr.select('a')[0])
#     print(tr.select("td"))
#     print(tr.select("td")[1])    
#     print(tr.select("td")[2])

    name = tr.select('a')[0].text
    departureTime = tr.select("td")[1].text
    arrivalTime   = tr.select("td")[2].text
    
    print("Name : ",name)
    print("Departure time : ",departureTime)    
    print("Arrival time : ",arrivalTime)
    
    data = {
        "name" : name,
        "departureTime" : departureTime,
        "arrivalTime"   : arrivalTime
    }
    dataList.append(data)

    print("-"*80)
```

```python
# 轉成 Dataframe
import pandas as pd
df = pd.DataFrame(dataList)
df = df[["name","departureTime","arrivalTime"]]
df.columns = ["車種","出發時間","抵達時間"]
df
```

```python
with open("sample-crawler-1.txt","w") as outFile:
    headerMsg = "車種 出發時間 抵達時間"
    outFile.write(headerMsg)
    outFile.write("\n")
    for ele in dataList:
        msg = "{} {} {}".format(ele["name"],ele["departureTime"],ele["arrivalTime"])
        outFile.write(msg)
        outFile.write("\n")
        
```
